This is the implements Gaussian elimination to solve a system of linear equations.

The system of equations as an augmented matrix, transforms it into an upper triangular form using elementary row operations (Gaussian elimination), and then solves for the variables using back substitution.



**Input:** The code first takes the number of rows and columns for the coefficient matrix and the elements of the matrix from the user.

```python
rows = int(input("Enter the number of rows"))
col = int(input("Enter the number of cols"))

```

```python
mat = []
print("Enter elements rows by rows")
```

```python
for i in range(rows):
  row = list(map(int,input().split()))
  mat.append(row)
```

check_square_mat(m) function: This function checks if the input coefficient matrix is a square matrix (number of rows equals number of columns). Gaussian elimination is typically applied to square matrices.

```python
def check_square_mat(m):
    no_of_rows = len(m)
    for i in mat:
      length = len(i);
      if(length != no_of_rows):
         return False;
    return True
```

check_whether_both_mat_equal(mat,colum_mat) function: This function checks if the number of rows in the coefficient matrix is equal to the number of elements in the constant list. This is necessary to form an augmented matrix.

```python
def check_whether_both_mat_equal(mat,colum_mat):
  return len(mat) == len(colum_mat)
```

```python
check_whether_both_mat_equal(mat,colum_mat)
```

make_it_agumented(mat,colum_mat) function: This function creates the augmented matrix by combining the coefficient matrix and the constant list. Each row of the augmented matrix contains a row from the coefficient matrix followed by the corresponding constant term.

```python
def make_it_augmented(mat,colum_mat):
  augmented_mat = [];
  for i in range(rows):
    augmented_mat.append(mat[i] + [colum_mat[i]])
  return augmented_mat
```

```python
argMat = make_it_augmented(mat,colum_mat)
```

```python
for i in argMat:
  print(i)
```

eleminate_col(augmented_mat, col)function: This function performs the elimination step for a specific column (col). It aims to make the elements below the pivot element (the diagonal element in the current column) zero using row operations

```python
def eleminate_col(augmented_mat,col):
  target_pivot = augmented_mat[col][col]
  n = len(augmented_mat)
  if target_pivot == 0:
      return augmented_mat

  for i in range(col+1,n):
      factor = augmented_mat[i][col] / target_pivot
      for j in range(len(augmented_mat[i])):
            augmented_mat[i][j] = augmented_mat[i][j] - factor * augmented_mat[col][j]
  return augmented_mat
```

convert_mat_into_upperTriangular(augmented_mat) function: This function iterates through the columns and applies the eleminate_col function to transform the augmented matrix into an upper triangular matrix

```python
def convert_mat_into_upperTriangular(augmented_mat):
  for i in range(len(augmented_mat)):
    augmented_mat = eleminate_col(augmented_mat,i)
  return augmented_mat
```

```python
resultArgMat = convert_mat_into_upperTriangular(argMat)
```

```python
for ele in resultArgMat:
  print(ele)
```

subsition(resultAugmentedMat) function: This function performs back substitution on the upper triangular augmented matrix to find the solution to the system of linear equations. It starts from the last row and solves for the variables.

```python
def subsition(resultAugmentedMat):
  n = len(resultAugmentedMat)
  m = len(resultAugmentedMat[0])
  result = [0]*n;
  for i in range(n-1,-1,-1):
    rhs = resultAugmentedMat[i][m-1]
    for j in range(i+1,n):
      rhs = rhs - resultAugmentedMat[i][j]*result[j]
    result[i] = rhs/resultAugmentedMat[i][i]
  return result
```

```python
res = subsition(resultArgMat)
```

```python
for r in res:
  print(r)
```
